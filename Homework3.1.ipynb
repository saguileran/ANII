# 1.T

Let $p$ be a positive constant. Deduce the a priori error estimation $ \big($ on the norm  $ H^1: ||e||^2_{H^1}=||e'||^2_{L^2}+||e||^2_{L^2} \big) $ if a 
picewise linear continous functions finite element method is used to the following problem

$$
-u''+pxu'+\left( 1+\frac{p}{2}\right)u = f, \quad \text{on } (0,1),\\
u(0)=0, \qquad u(1) = 0
$$

 ## Solution

We multiply the differential ecuation by a test function $v \in H^1(0,1)$ , and integrate over $(0,1)$. Using partial integration and the boundary conditions we get the following variational problem:

Find $u \in H^1(0,1)$ such that

$\int_{0}^{1}(u'v'+pxu'v +\frac{p}{2}uv +uv)dx = \int_{0}^{1}fvdx, \forall v \in V_h^{0} \subset H^1(0,1)$ $\quad  (1)$


Finite Element Method :
Find $U \in V_h^{0}$ such that $\int_{0}^{1}(U'v'+pxU'v +\frac{p}{2}Uv +Uv)dx = \int_{0}^{1}fvdx,  \forall v \in V_h^{0} \subset H^1(0,1)$ $\quad  (2)$

where  $V_h^{0}=\{$ $v$ :$v$ is piecewise linear and continuous in a partition of $(0,1)$, $v(0) = v(1) = 0$$\}$

Now let $e = u − U$, then $(1)-(2)$ gives that $\int_{0}^{1}(e'v'+pxe'v +\frac{p}{2}ev +ev)dx = 0,  \forall v \in V_h^{0}$ $\quad  (3)$

We note that using $e(0) = e(1) = 0$, we get

$\int_{0}^{1} pxe'e dx +\int_{0}^{1}\frac{p}{2}ee dx =\frac{p}{2}\int_{0}^{1}x\frac{d(e^2)}{dx}+\int_{0}^{1}\frac{p}{2}e^2 dx=\frac{p}{2}(xe^2)\bigg|_{0}^{1}-\frac{p}{2}\int_{0}^{1}e^2dx+\frac{p}{2}\int_{0}^{1}e^2dx = 0$ $\quad  (4)$

A priori error estimate: We use $(4)$ and write:
\begin{align*}
||e||_{H^{1}}^2 = \int_{0}^{1}(e'e'+ee)dx &=\int_{0}^{1}(e'e'+pxe'e+\frac{p}{2}ee+ee)dx \\
&=\int_{0}^{1}e'(u-U)'+pxe'(u-U)+\frac{p}{2}e(u-U)+e(u-U)dx = \{v= u-\pi u_h \quad in \quad (3)\} \\
&=\int_{0}^{1}e'(u-\pi u_h)'+pxe'(u-\pi u_h)+\frac{p}{2}e(u-\pi u_h)+e(u-\pi u_h)dx \\
&\leq ||e'|| \ ||(u-\pi u_h)'||+p||e'|| \ ||u-\pi u_h||+\frac{p}{2}||e|| \ ||u-\pi u_h||+||e|| \ ||u-\pi u_h|| \\
&\leq \{||(u-\pi u_h)'||+(p+1)||u-\pi u_h||\}||e||_{H^{1}} \\
&\leq C_i\{||hu''||+||h^2u''||\}||e||_{H^{1}} \\
\end{align*}

where in the last step we used Poincare inequality. This gives that

$$||e||_{H^{1}}\leq C_i\{||hu''||+||h^2u''||\}$$

which is the a priori error estimate.


# 2.T
Consider the problem

$$
-\varepsilon u'' + c(x) u' +u = f(x), \qquad 0<x<1\\
u(0)=u'(1)=0
$$

where $ \varepsilon $ is a positive constant, the function $ c $ satisfies $ c(x)\geq 0, \; c'(x)\leq 0, $ and $ f\in L^2 (0,1) $. Prove that there are some positive constants $ C_1, C_2, $ and $ C_3 $ such that

$$
\sqrt{\varepsilon} ||u'| \leq C_1 ||f||, \qquad ||cu'|| \leq C_2||f||, \qquad \text{and} \qquad ||u'|| \leq C_3 ||f||
$$

where $ || \cdot || $ is the $ L^2(0,1) $ norm.

## Solution

In general, a good idea is to multiply by u and integrate by parts, originally you have the equation:

$$
 -\epsilon u'' + c(x)u'+ u = f \in [0,1]
$$
And the  boundary conditions $u(0) = u'(1) = 0$.

Before integrating, it is possible to say that for the spaces  $L^2$
\begin{align}
    \lVert\epsilon u'' \rVert + \lVert f \rVert = \lVert c(x)u'+ u \rVert \ge \lVert c(x) u' \rVert\\
    \epsilon\lVert f \rVert \ge \epsilon\lVert u\rVert\\
    (\epsilon+a)\lVert f \rVert \ge \epsilon\lVert u\rVert\\
    C_1\lVert f \rVert \ge \epsilon\lVert u\rVert\\
\end{align}

This in itself already satisfies the first desired inequality, additionally multiplying by u to both parts and applying integration by parts
\begin{align}
   \epsilon \int_{0}^{1}(u')^{2}dx + \int_{0}^{1}c(x)u'u + \int_{0}^{1} u^2 = \int_{0}^{1} fu
\end{align}

Therefore

\begin{align}
   \epsilon ||u'||^2 + \int_{0}^{1}c(x)u'u + ||u||^2 = \int_{0}^{1} f(u) \leq ||f|| \ ||u||
\end{align}

If we apply the Cauchy Schwarz inequality then 
\begin{align}
    \int_{0}^{1}c(x)u'u dx  \leq ||cu'||_{L^{2}} \ ||u||_{L^{2}}
\end{align}

Since $c'(x) \leq 0$, then $c$ reaches a maximum at the left boundary of the domain, at $x = 0$ and since $c(x)\leq 0$, in the whole domain, its maximum, which we can call $c(0)$, is positive, and thus 

\begin{align}
     ||cu'||_{L^{2}} \ ||u||_{L^{2}} \leq c(0) ||u'||_{L^{2}}  ||u||_{L^{2}}
\end{align}

Then,
\begin{align}
     ||cu'||_{L^{2}}  \leq c(0) ||u'||_{L^{2}} \leq \frac{c(0)}{\epsilon} ||f||_{L^{2}}
\end{align}

Thus concluding that 
\begin{align}
     ||cu'||_{L^{2}} \leq \frac{c(0)}{\epsilon} ||f||_{L^{2}}
\end{align}

Now, let's go back to the original equation for the first inequality, that is,
\begin{align}
 -\epsilon u'' + c(x)u'+ u = f 
\end{align}
Clearing and reorganizing
\begin{align}
 \epsilon u''  = c(x)u'+ u -f  
\end{align}
Finally, for the last inequality,
\begin{align}
 ||\epsilon u'' &||_{L^{2}} = ||c(x)u'+ u -f ||_{L^{2}}\\
 & \leq ||f||_{L^{2}} + ||cu'||_{L^{2}} + ||u||_{L^{2}}\\
 & \leq ||f||_{L^{2}} + \frac{c(0)}{\epsilon}||f||_{L^{2}} + ||f||_{L^{2}} \\
 &= \bigg( 2 + \frac{c(0)}{\epsilon}\bigg)||f||_{L^{2}}
\end{align}
This verifies the last desired inequality and successfully completes the proof.

### a)

Remembering the picewise lineal polynomial functions

$$
\phi_{i}(x):= \begin{cases}\frac{x-x_{i-1}}{h_{i}} & \text { if } x \in \Omega_{i} \\ \frac{x_{i+1}-x}{h_{i+1}} & \text { if } x \in \Omega_{i+1} \\ 0 & \text {other case}\end{cases}, \qquad \phi_{i}'(x):= \begin{cases}\frac{1}{h_{i}} & \text { if } x \in \Omega_{i} \\ \frac{-1}{h_{i+1}} & \text { if } x \in \Omega_{i+1} \\ 0 & \text {other case}\end{cases}
$$

let's calculate the weak form of the equation. Now, multiplying by a test function $v$ such that $v(0)=0$ and $v(1)=1$, itegrating over $[0,1]$ and applying integration by steps

$$ 
-\varepsilon \int_0^1 v u_{xx} dx + \int_0^1 v u_x dx = 0\\
\varepsilon \int_0^1 v' u' dx - \varepsilon vu'\Big|_0^1 + \int_0^1 v u' dx = \varepsilon \int_0^1 v' u' dx + \varepsilon[v(0)u'(0) - v(1)u'(1)] + \int_0^1 v u' dx = 0\\
\varepsilon \int_0^1 v' u' dx  + \int_0^1 v u' dx = \varepsilon v(1)u'(1) =\varepsilon u'(1)
$$

Using the picewise linear polynomials

$$ 
u_h(x) = \sum_{j=0}^{n} u_j \phi_j(x), \qquad u_h'(x) = \sum_{j=0}^{n} u_j \phi_j'(x), \qquad v(x) = \phi_i(x)

$$

\sum_{j=0}^{n} \left[\varepsilon \int_0^1 \phi_i'(x) \phi_j'(x) dx  +  \int_0^1 \phi_i(x) \phi_j'(x) dx \right] u_j = \varepsilon \phi_n'(1) = \frac{\varepsilon}{h_n}
$$

That can be writen as a matricial form $(A+M)u=b$ with

$$
A = (a_{ij}) = \epsilon \int_0^1  \phi_i'(x) \phi_j'(x) dx\\
M = (m_{ij}) =   \int_0^1 \phi_i(x) \phi_j'(x) dx 
$$


- if $j=i-1$

$$
a_{i, i-1} = \epsilon \int_0^1  \phi_i' \phi_{i-1}' dx = \epsilon \int_{x_{i-1}}^{x_i}  \frac{1}{h_i} \; \cdot  \; \frac{-1}{h_i} dx = -\frac{\epsilon}{h_i^2}h_i = -\frac{\epsilon}{h_i}\\
m_{i, i-1} =  \int_0^1 \phi_i \phi_{i-1}' dx = \int_{x_{i-1}}^{x_i} \left( \frac{x-x_{i-1}}{h_i}\right) \; \cdot  \; \frac{-1}{h_i} dx = - \frac{1}{h_i^2}\left(\frac{1}{2}x^2-x_{i-1}x\right)\Big|_{x_{i-1}}^{x_i}\\
= - \frac{1}{h_i^2}\left(\frac{1}{2}x_i^2-x_{i-1}x_i - \frac{1}{2}x_{i-1}^2+x_{i-1}^2\right) = - \frac{1}{h_i^2}\frac{h_i^2}{2} = -\frac{1}{2}
$$

- if $j=i$

$$
a_{i, i} = \epsilon \int_0^1  \phi_i' \phi_{i}' dx =  \epsilon \int_{x_{i-1}}^{x_{i+1}}  \phi_i' \phi_{i}' dx = \epsilon \int_{x_{i-1}}^{x_i}  \phi_i' \phi_{i}' dx + \epsilon \int_{x_{i}}^{x_{i+1}}  \phi_i' \phi_{i}' dx \\
= \epsilon \int_{x_{i-1}}^{x_i}  \frac{1}{h_i^2} dx + \epsilon \int_{x_{i}}^{x_{i+1}} \frac{1}{h_{i+1}^2} dx  = \epsilon\left(\frac{1}{h_i} + \frac{1}{h_{i+1}}\right) = \frac{\epsilon}{h_{i} h_{i+1}} (h_{i+1} + h_{i})\\
m_{i, i} =  \int_0^1 \phi_i \phi_{i}' dx = \int_{x_{i-1}}^{x_{i+1}}  \phi_i \phi_{i}' dx = \int_{x_{i-1}}^{x_{i}}  \phi_i \phi_{i}' dx + \int_{x_{i}}^{x_{i+1}}  \phi_i \phi_{i}' dx \\
\int_{x_{i-1}}^{x_{i}}   \left(\frac{x-x_{i-1}}{h_i} \right) \frac{1}{h_i} dx + \int_{x_{i}}^{x_{i+1}}   \left(\frac{x_{i+1}-x}{h_{i+1}} \right) \frac{-1}{h_{i+1}}  dx = \frac{1}{h_i^2}  \left(\frac{1}{2}x^2-x_{i-1}x\right)\Big|_{x_{i-1}}^{x_{i}} - \frac{1}{h_{i+1}^2}\left(x_{i+1}x-\frac{1}{2}x^2\right)\Big|_{x_{i}}^{x_{i+1}}\\
= \frac{1}{h_i^2} \frac{h_i^2}{2}- \frac{1}{h_{i+1}^2}\frac{h_{i+1}^2}{2} = 0
$$


Thus, the matrices and vector are

$$
\boxed{
M = -\frac{1}{2}\begin{pmatrix}

& \ddots & \ddots & \ddots \\

```julia
function Stiffness1D_Hom(x, epsilon)
    n = size(x,1)-2;    #x0< x1 < ... < xN n+1 puntos en la malla %
    A = zeros(n,n);
    A[1,1], A[n,n]= 1/(x[2]- x[1]) , 1/(x[end] - x[end-1])
    [A[i-1:i,i-1:i] += 1/(x[i+1] - x[i]) * [1 -1; -1 1] for i in 2:n]
    
    return epsilon .* A
end

function Mass1D_Hom(x)
    n = size(x,1)-2;
    M = zeros(n,n);  
    #M[1,1], M[n,n] =  r( 0.5*(x[2]+ x[1]) )* ( x[2]- x[1])/3, r( 0.5*(x[n+2]+ x[n+1]) )*  ( x[n+2] - x[n+1])/3;
    [M[i-1:i,i-1:i] += [0 1; 1 0] for i in 2:n ]
    
    return -0.5 .* M
end

#x = LinRange(0,1,1000)
#Stiffness1D_Hom(x, 0.1)
#Mass1D_Hom(x)
```

### b)

Remembering the exact solution to the case where $\beta = 1 >> \varepsilon$,

$$
u(x)\approx \frac{e^{\beta/\varepsilon x -1 }}{e^{\beta/\varepsilon x -1}} = e^{\beta/\varepsilon (x -1)}
$$

```julia

function Solution(par, Stiffness)
    n, epsilon = par
    x = LinRange(0,1,n)

    M = Mass1D_Hom(x)
    A = Stiffness(x, epsilon)
    
    b = zeros(n-2,1);
    b[end] = epsilon / (x[end]-x[end-1])
    
    
    usol = [0; (M+A) \ b; 1]
    return x, usol
end

function SOL(Stiffness)
    X, U = zeros(0), zeros(0)

    for i in 1:1:size(parameters,1)
        x, usol = Solution(parameters[i], Stiffness)
        X, U = vcat(X, [x]), vcat(U, [usol])
    end
    return X, U
end;
```

### c)

To solve correctly the problem adaptative methods are used, since $\varepsilon << 1$ the advection dominates over the difusion. To solve the present inestabilities let's remember the definition of global Peclet number

$$
\mathbb{P}e_g = \frac{|\beta|L}{2\varepsilon} =  \frac{1}{2\varepsilon}
$$

Using the previous number we can define a viscosity coefficient as

$$
\varepsilon_h =  \varepsilon(1 + \phi(\mathbb{P}e))
$$

where $\phi$ is function of the global Peclet number such that

$$
\lim_{t \to O^+} \phi(t) = 0 
$$

note that when $\phi=0$ we get the original problem.

#### Numerical Viscosity (Upwind)

Taking $\phi(\mathbb{P}e) = \mathbb{P}e$, then 

$$\varepsilon_h =  \varepsilon(1 + \mathbb{P}e)  =  \varepsilon\left(1 + \frac{h_i}{2\varepsilon}\right)$$

$$
\boxed{
A = \varepsilon\begin{pmatrix}
\left(1 + \frac{h_1}{2\varepsilon}\right) \left(\frac{1}{h_1} + \frac{1}{h_{2}}\right)  & -\frac{1}{h_2} \left(1 + \frac{h_1}{2\varepsilon}\right) & 0 & \\
-\frac{1}{h_2} \left(1 + \frac{h_2}{2\varepsilon}\right) & \left(1 + \frac{h_2}{2\varepsilon}\right)\left(\frac{1}{h_2} + \frac{1}{h_{3}}\right)  & -\frac{1}{h_3} \left(1 + \frac{h_2}{2\varepsilon}\right) & 0 &\\
& \ddots & \ddots & \ddots \\
&0 & -\frac{1}{h_{n-2}} \left(1 + \frac{h_{n-2}}{2\varepsilon}\right)& \left(1 + \frac{h_{n-2}}{2\varepsilon}\right) \left(\frac{1}{h_{n-2}} + \frac{1}{h_{n-1}}\right)  & -\frac{1}{h_{n-1}} \left(1 + \frac{h_{n-2}}{2\varepsilon}\right) \\
 && 0 & -\frac{1}{h_{n-1}} \left(1 + \frac{h_{n-1}}{2\varepsilon}\right) & \left(1 + \frac{h_{n-1}}{2\varepsilon}\right) \left(\frac{1}{h_{n-1}} + \frac{1}{h_{n}}\right)  
\end{pmatrix}}
$$

```julia
function Stiffness1D_Hom_Vis(x, epsilon)
    n = size(x,1)-2;    #x0< x1 < ... < xN n+1 puntos en la malla %
    A = zeros(n,n);
    A[1,1], A[n,n]= (1+(x[2]- x[1])/(2*epsilon)) / (x[2]- x[1]) , (1+(x[n+2]- x[n+1])/(2*epsilon)) / ( x[n+2] - x[n+1])
    [A[i-1:i,i-1:i] += 1/(x[i+1] - x[i]) * [(1+(x[i]- x[i-1])/(2*epsilon)) -(1+(x[i]- x[i-1])/(2*epsilon)); 
                                           -(1+(x[i+1]- x[i])/(2*epsilon))  (1+(x[i+1]- x[i])/(2*epsilon))] for i in 2:n]
    
    return epsilon .* A
end
```

#### Optimal Viscosity (Scharfetter - Gummel)

Taking $\phi(\mathbb{P}e) = \mathbb{P}e-1+B(2\mathbb{P}e)$, where

$$
B({\mathbb{P}e}) = \begin{cases} \frac{\mathbb{P}e}{e^{\mathbb{P}e}-1} & \text { if } \mathbb{P}e > 0 \\ 1 & \text { if } \mathbb{P}e = 0 \end{cases}
$$

then, since the Peclet number is always bigger than $0$.

$$\varepsilon_h =  \varepsilon(\mathbb{P}e+B(2\mathbb{P}e)) = \varepsilon\left(\frac{h_i}{2\varepsilon}+\frac{h_i}{\varepsilon} \frac{1}{e^{\frac{h_i}{\varepsilon}}-1}\right) = \frac{h_i}{2}\left(1+ \frac{2}{e^{\frac{h_i}{\varepsilon}}-1}\right) $$

$$
A = \begin{pmatrix}
\frac{h_1}{2}\left(1+ \frac{2}{e^{\frac{h_1}{\varepsilon}}-1}\right) \left(\frac{1}{h_1} + \frac{1}{h_{2}}\right)  & -\frac{1}{h_2} \frac{h_1}{2}\left(1+ \frac{2}{e^{\frac{h_i}{\varepsilon}}-1}\right) & 0 & \\
-\frac{1}{h_2} \frac{h_2}{2}\left(1+ \frac{2}{e^{\frac{h_2}{\varepsilon}}-1}\right) & \frac{h_2}{2}\left(1+ \frac{2}{e^{\frac{h_2}{\varepsilon}}-1}\right) \left(\frac{1}{h_2} + \frac{1}{h_{3}}\right)  & -\frac{1}{h_3} \frac{h_2}{2}\left(1+ \frac{2}{e^{\frac{h_2}{\varepsilon}}-1}\right) & 0 &\\
& \ddots & \ddots & \ddots \\
&0 & -\frac{1}{h_{n-2}} \frac{h_{n-2}}{2}\left(1+ \frac{2}{e^{\frac{h_{n-2}}{\varepsilon}}-1}\right) & \frac{h_{n-2}}{2}\left(1+ \frac{2}{e^{\frac{h_{n-2}}{\varepsilon}}-1}\right) \left(\frac{1}{h_{n-2}} + \frac{1}{h_{n-1}}\right)  & -\frac{1}{h_{n-1}} \frac{h_{n-2}}{2}\left(1+ \frac{2}{e^{\frac{h_{n-2}}{\varepsilon}}-1}\right) \\
 && 0 & -\frac{1}{h_{n-1}} \frac{h_{n-1}}{2}\left(1+ \frac{2}{e^{\frac{h_{n-1}}{\varepsilon}}-1}\right) & \frac{h_{n-1}}{2}\left(1+ \frac{2}{e^{\frac{h_{n-1}}{\varepsilon}}-1}\right) \left(\frac{1}{h_{n-1}} + \frac{1}{h_{n}}\right)

```julia
function peclet(p,h)
    return (h/2*p(h))
end

function ensamble_rigidez(x,f)
    n = length(x)-2
    A = np.zeros([n,n])  
    A[0,0]= f( 0.5*(x[1]+ x[0])) / ( x[1]- x[0])  
    A[n-1,n-1]= f( 0.5*(x[n+1]+ x[n]) )/ ( x[n+1]- x[n]) 
    print(peclet(f,x[1]- x[0]))
    for j in np.arange(1,n)
        h = x[j+1] - x[j]
        xm = (x[j+1] + x[j])/2
        fm = f(xm)
        A[j-1,j-1] +=   fm/h     #% sum pmid/h a A(i-1,i-1)
        A[j-1,j] += - fm/h  
        A[j,j-1] += - fm/h   
        A[j,j] += fm/h   
    end   
    return A
end  
```

```julia
function ensamble_masa(x,r)
    n = len(x)-2
    M = np.zeros([n,n])
    M[0,0]=  r( 0.5*(x[1]+ x[0]) )/ ( x[1]- x[0])
    M[n-1,n-1]=  r( 0.5*(x[n+1]+ x[n]) )/ ( x[n+1] - x[n])
    

    for i in np.arange(1,n)
        h = x[i+1] - x[i]
        xmid = (x[i+1] + x[i])/2
        rmid = r(xmid)

        M[i-1,i-1] = 0#M[i-1,i-1] +  rmid*h/3
        M[i-1,i] = 0.5#*(M[i-1,i] +  rmid*h/6)
        M[i,i-1] = 0.5#*(M[i,i-1] + rmid*h/6)
        M[i,i] = 0#M[i,i] + rmid*h/3
    end
    return M 
end
```

```julia
function ensamble_carga(x,g)
    n = len(x)-2
    b = np.zeros([n,1])
    b[0]= g(x[0])*(x[1]- x[0])
    b[n-1]= g(x[n+1])*(x[n+1]- x[n])
    for i in np.arange(1,n)
        h = x[i+1] - x[i]
        xmid = (x[i+1] + x[i])/2
        b[i] += g(xmid)*h
        b[i-1] += g(x[i-1])*h
    end
    return b
end
```

```julia
function error_residual(x,f,u,q,r)
    n = len(x)-1
    eta = np.zeros([n])
    #print(eta)
    for k in range(n)
        h = x[k+1] - x[k]
        a = f(x[k]) - q(x[k])*(u[k+1]-u[k])/h + r(x[k])*u[k]
        b = f(x[k+1]) - q(x[k+1])*(u[k+1]-u[k])/h + r(x[k+1])*u[k+1]
        t = (a^2+b^2)*h/2
        eta[k] = h^4*t
    end
    return eta
end
```
