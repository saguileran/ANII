```julia
function Stiffness1D_Hom(x,p)
    # Stiffness matrix for homogeneous Dirichlet problem.
    # x = LinRange(0,1,11) ;  
    # p(x) = x^3-0.18*x^2 + 691e-6;
    # A = EnsambleRigidez1D_Hom(x,p);

    n = size(x)[1]-2;    #x0< x1 < ... < xN n+1 puntos en la malla %
    A = zeros(n,n);
    A[1,1]= p(0.5*(x[2]+ x[1])) / (x[2]- x[1]);
    A[n,n]= p( 0.5*(x[n+2]+ x[n+1]) )/ ( x[n+2] - x[n+1]);
    for i in 2:n
        h = x[i+1] - x[i];
        xmid = (x[i+1] + x[i]) / 2;       # midpoint of interval
        pmid = p(xmid);                   # p(x) value in the midpoint
        A[i-1,i-1] = A[i-1,i-1] + pmid/h; # sum pmid/h a A(i-1,i-1)
        A[i-1,i]   = A[i-1,i]   - pmid/h;
        A[i,i-1]   = A[i  ,i-1] - pmid/h;
        A[i,i]     = A[i  ,i]   + pmid/h;
    end
    return A
end

x    = 0:0.1:1;  
p(x) = x^3 - 0.18*x^2 + 691e-6;
A    = Stiffness1D_Hom(x,p)
A
```

```julia
function Load1D_hom(x,f)
    # Load vector homogeneous Dirichlet problem.
    # x = LinRange(0,1,11);
    # f(x) = x^3-0.18*x^2 + 691e-6);
    # b = EnsambleCarga1D_hom(x,f);

    n    = size(x,1)-2;
    b    = zeros(n,1);
    b[1] = 0.5* f(x[1])* (x[2]- x[1]);
    b[n] = 0.5* f(x[n+2])* (x[n+2]- x[n+1]);
    for i in 2:n
        h      = x[i+1] - x[i];
        b[i-1] = b[i-1] + f(x[i-1])*h/2;
        b[i]   = b[i] + f(x[i])*h/2;
    end
    return b
end

x    = LinRange(0,1,11);
f(x) = x^3-0.18*x^2 + 691e-6;
b    = Load1D_hom(x,f);
b
```

```julia
function Mass1D_Hom(x,r)
    # Mass matrix for homogeneous Dirichlet problem.
    # x    = LinRange(0,1,11);
    # r(x) = x^3-0.18*x^2 + 691e-6
    # M    = EnsambleMasa1D_Hom(x,r);

    n = size(x,1)-2;
    M = zeros(n,n);  
    M[1,1] =  r( 0.5*(x[2]+ x[1]) )* ( x[2]- x[1])/3;
    M[n,n] =  r( 0.5*(x[n+2]+ x[n+1]) )*  ( x[n+2] - x[n+1])/3;
    for i in 2:n 
        h = x[i+1] - x[i];        # Interval lenght
        xmid = (x[i+1] + x[i])/2; # midpoint of interval
        rmid = r(xmid);           # r(x) value to midpoint of interval
        M[i-1,i-1] = M[i-1,i-1] + rmid*h/3; # sum rmid*h/3 a M(i,i)
        M[i-1,i]   = M[i-1,i]   + rmid*h/6;
        M[i  ,i-1] = M[i,  i-1] + rmid*h/6;
        M[i  ,i]   = M[i,  i]   + rmid*h/3;
    end
    return M
end

x    = LinRange(0,1,11);
r(x) = x^3-0.18*x^2 + 691e-6;
M    = Mass1D_Hom(x,r);
M
```

# Example

$$ \eta_i = h_i ||f+u_h''|| \qquad, \text{with} \qquad u_h=0 \qquad \text{()}$$,

$$-u'' = f=e^{-100|x-0.5|^2}\\
u(0)=0, \qquad u(1)=0$$

```julia
function Refine(Ruh, x)
    alpha = 0.9
    for i=1:size(Ruh,1)
       if Ruh[i] > alpha*maximum(Ruh)
            x = [x; (x[i]+x[i+1])/2]     
       end
    end
    return sort(x)
end

function ResiudalError(x,f) #Ruh
    n = size(x,1)-1
    Ruh = zeros((n,1))
    for i=1:n
        h = x[i+1] - x[i]
        a, b = f(x[i]), f(x[i+1])
        t = (a^2 + b^2) * h/2
        Ruh[i] = h*sqrt(t)
    end
    return Ruh
end
```

```julia
using PlotlyJS, LinearAlgebra
```

$$ -(pu')'+u'=f, \quad u=u(x)$$


```julia
# Creating Mesh
function SolveEq(N, p, r, f)
    #a, b = 0, 1;
    #N = 500
    #h = (b-a) / N
    
    h = 1/N
    x = 0:h:1

    # Linear system
    A  = Stiffness1D_Hom(x,p);
    M  = Mass1D_Hom(x,r);
    bc = Load1D_hom(x,f);

    # ------  Solution
    ua, ub = 0, 0; #Dirichle homogeneous conditions
    sol = [ua; (A + M) \ bc; ub]

    Nmax = 10
    n = 1
    # Refinamiento
    x_cop = copy(x)
    while n < Nmax #Ruhx

        Ruh = ResiudalError(x_cop,f) 
        x_cop  = Refine(Ruh, x_cop)

        A  = Stiffness1D_Hom(x_cop,p);
        M  = Mass1D_Hom(x_cop,r);
        bc = Load1D_hom(x_cop,f);

        global solx = [ua; (A + M) \ bc; ub]

        n = n+1
    end
    return x, sol, x_cop, solx 
end

# Problem functions ((pu')'+qu'+ru=f, u=u(x))
c = 100
p(x) = 1
r(x) = 0
f(x) = exp(-c*abs(x-0.5)^2)

x, sol, x_cop, solx = SolveEq(10, p, r, f);
x, sol, x_cop1, solx1 = SolveEq(20, p, r, f);
x, sol, x_cop2, solx2 = SolveEq(50, p, r, f);
x, sol, x_cop3, solx3 = SolveEq(100, p, r, f);
```

```julia
plot([scatter(x=Array(x_cop),y=solx[:,1],mode="markers", name="Refined N=10"), 
        scatter(x=Array(x_cop1),y=solx1[:,1],mode="markers", name="Refined N=20"), 
        scatter(x=Array(x_cop2),y=solx2[:,1],mode="markers", name="Refined N=50"), 
        scatter(x=Array(x_cop3),y=solx3[:,1],mode="markers", name="Refined N=100"), 
        scatter(x=Array(x),y=sol[:,1],mode="markers", name="Equidistant")])
```

$$-\mu u'' + \sigma u = 0 \\
u(0) = 0, \qquad u(1)=1$$

with $\mu = 10$ and $\sigma = 12000$. With exact solution 
$$u_{exact} (x) = \frac{e^{\alpha x} - e^{-\alpha x}}{e^{\alpha} - e^{-\alpha}}$$

To solve this equation, it is necessary a sustitution to do both boundary conditions be zero, $u = w+y$ and making $y=x$ $(y(0)=0,\; y(1)=1)$ such that satisfy the $u$ boundary conditions

$$-\mu w'' + \sigma w = - \sigma x \\
w(0) = 0, \qquad w(1)=0$$



```julia
mu, sigma = 10, 12e3

p(x) = mu
r(x) = sigma  # sigma
f(x) = - sigma * x

x, sol, x_cop,  solx  = SolveEq(10, p, r, f);
x, sol, x_cop1, solx1 = SolveEq(20, p, r, f);
x, sol, x_cop2, solx2 = SolveEq(50, p, r, f);
x, sol, x_cop3, solx3 = SolveEq(100, p, r, f);
```
